```python
import sqlalchemy as db
engine = db.create_engine('postgresql://localhost/portfolio_db', echo=False)
connection = engine.connect()
metadata = db.MetaData()
```

```python
portfolio = db.Table('portfolio', metadata, autoload=True, autoload_with=engine)
```

```python
print(portfolio.columns.keys())
```

```python
# Print full table metadata
print(repr(metadata.tables['portfolio']))
```

```python
query = db.select([portfolio])
```

```python
ResultProxy = connection.execute(query)
```

```python
ResultSet = ResultProxy.fetchall()
```

```python
ResultSet[:3]
```

```python
import pandas as pd
df = pd.DataFrame(ResultSet)
df.columns = ResultSet[0].keys()
```

```python
df.head()
```

```python
query_gb = (db.select([portfolio.columns.travel_product,
                       db.func.sum(portfolio.columns.spend_ac).label('spend_ac'), 
                      db.func.sum(portfolio.columns.bookings_ac).label('bookings_ac')
                      ])
            .group_by(portfolio.columns.travel_product))
```

```python
ResultProxy_gb = connection.execute(query_gb)
```

```python
ResultSet_gb = ResultProxy_gb.fetchall()
```

```python
df_gb = pd.DataFrame(ResultSet_gb)
df_gb.columns = ResultSet_gb[0].keys()
```
